```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'penguins.csv')
data.head()
```

```python
data.shape
```

```python
X = data[['масса тела']]
y = data['длина плавника']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)
```

```python
plt.scatter(X_train, y_train)

plt.xlabel("масса тела в мм")
plt.ylabel("длина плавника в мм")

plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(X_train, y_train)

def f_line(x):
    return lr.coef_ * x + lr.intercept_

x = [2500, 6500]
y = f_line(x)

plt.plot(x, y)
plt.xlabel("масса тела в мм")
plt.ylabel("длина плавника в мм")

plt.show()
```

```python
plt.scatter(X_test, y_test)

def f_line(x):
    return lr.coef_ * x + lr.intercept_

x_line = [2500, 6500]
y_line = f_line(x)

plt.plot(x_line, y_line)
plt.xlabel("масса тела в мм")
plt.ylabel("длина плавника в мм")

plt.show()
```

```python
y_train_predict = lr.predict(X_train)

y_test_predict = lr.predict(X_test)
```

```python
np.array(y_test)
```

```python
np.array(y_test_predict)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(y_train_predict, y_train)
```

```python
mean_absolute_error(y_test_predict, y_test)
```

```python
mean_squared_error(y_train_predict, y_train)
```

```python
mean_squared_error(y_test_predict, y_test)
```

```python
X = data[['масса тела', 'длина клюва']]
y = data['длина плавника']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15)
```

```python
fig = plt.figure(figsize=(7,7))

ax = fig.add_subplot(projection='3d')

x_points= X_train['масса тела']
y_points = X_train['длина клюва']
z_points = y_train

ax.scatter(x_points, y_points, z_points)

ax.set_xlabel("масса тела в мм")
ax.set_ylabel("длина клюва в мм")
ax.set_zlabel("длина плавника в мм")

plt.show()
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_train_predict = lr.predict(X_train)
```

```python
y_train_predict
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
def f_plane(x1, x2):
    return lr.coef_[0] * x1 + lr.coef_[1] * x2 + lr.intercept_
```

```python
X_train['длина клюва'].min(), X_train['длина клюва'].max()
```

```python
fig = plt.figure(figsize=(7,7))

ax = fig.add_subplot(projection='3d')

x_points= X_train['масса тела']
y_points = X_train['длина клюва']
z_points = y_train

ax.scatter(x_points, y_points, z_points)

x, y  = np.meshgrid(range(2800, 6500), range(30, 61))
z = np.array([f_plane(point_x, point_y) for point_x, point_y in zip(x, y)])
ax.plot_surface(x, y, z, alpha=0.2)


ax.set_xlabel("масса тела в мм")
ax.set_ylabel("длина клюва в мм")
ax.set_zlabel("длина плавника в мм")

plt.show()
```
