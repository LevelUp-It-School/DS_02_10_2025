$$f(x_1, x_2) = x_1 x_2 + x_1^2$$



$w_1 = x_1$, 

$w_2 = x_2$, 

$w_3 = w_1w_2$, 

$w_4 = w_1^2$, 

$w_5 = w_3 + w_4$

$$\dfrac{\partial{f(x_1^*, x_2^*)}}{\partial{x_1}}$$


$$\dfrac{\partial{f}}{\partial{x_1}} = \dfrac{\partial{f}}{\partial{w_5}} \dfrac{\partial{w_5}}{\partial{x_1}} = \dfrac{\partial{f}}{\partial{w_5}} \Big[ \dfrac{\partial{w_5}}{\partial{w_4}} \dfrac{\partial{w_4}}{\partial{x_1}} + \dfrac{\partial{w_5}}{\partial{w_3}} \dfrac{\partial{w_3}}{\partial{x_1}} \Big] = \cdots$$


$$\dfrac{\partial{w_1(x_1^*, x_2^*)}}{\partial{x_1}} = 1$$

$$\dfrac{\partial{w_2(x_1^*, x_2^*)}}{\partial{x_1}} = 0$$

$$\dfrac{\partial{w_3(x_1^*, x_2^*)}}{\partial{x_1}} = \dfrac{\partial{w_1(x_1^*, x_2^*)}}{\partial{x_1}} w_2 + \dfrac{\partial{w_2(x_1^*, x_2^*)}}{\partial{x_1}} w_1 = x_2^*$$

$$\dfrac{\partial{w_4(x_1^*, x_2^*)}}{\partial{x_1}} = 2 w_1\dfrac{\partial{w_1(x_1^*, x_2^*)}}{\partial{x_1}} = 2 x_1^*$$

$$\dfrac{\partial{w_5(x_1^*, x_2^*)}}{\partial{x_1}} = \dfrac{\partial{w_3(x_1^*, x_2^*)}}{\partial{x_1}} + \dfrac{\partial{w_4(x_1^*, x_2^*)}}{\partial{x_1}} = x_2^* + 2 x_1^*$$

$$\dfrac{\partial{f(x_1^*, x_2^*)}}{\partial{x_1}} = \dfrac{\partial{f(x_1^*, x_2^*)}}{\partial{w_5}} \dfrac{\partial{w_5(x_1^*, x_2^*)}}{\partial{x_1}} = x_2^* + 2 x_1^*$$

```python
class Variable:
    def __init__(self, value, derivative):
        self.value = value
        self.derivative = derivative

    def __add__(self, other):
        return Variable(
            self.value + other.value,
            self.derivative + other.derivative)
    def __mul__(self, other):
        return Variable(
            self.value * other.value,
            self.derivative * other.derivative + self.value * other.derivative)
    def __pow__(self, other):
        return Variable(
            self.value * other,
            other * self.value ** (other-1))

    


```

```python
def f(x1, x2):
    vx1 = Variable(x1, 1)
    vx2 = Variable(x2, 0)
    vf = vx1 * vx2 + vx1 ** 2
    return vf.value, vf.derivative

print(f(2, 4))
```

```python
car_prices_array = [3,4,5,6,7,8,9]
car_price_np = np.array(car_prices_array,dtype=np.float32)
car_price_np = car_price_np.reshape(-1,1)
car_price_tensor = Variable(torch.from_numpy(car_price_np))
```

```python
number_of_car_sell_array = [ 7.5, 7, 6.5, 6.0, 5.5, 5.0, 4.5]
number_of_car_sell_np = np.array(number_of_car_sell_array,dtype=np.float32)
number_of_car_sell_np = number_of_car_sell_np.reshape(-1,1)
number_of_car_sell_tensor = Variable(torch.from_numpy(number_of_car_sell_np))
```

```python
import matplotlib.pyplot as plt
plt.scatter(car_prices_array,number_of_car_sell_array)
plt.xlabel("   $")
plt.ylabel("-  ")
plt.show()
```

```python
car_prices = [3, 4, 5, 6, 7, 8, 9]
car_sales = [7.5, 7.0, 6.5, 6.0, 5.5, 5.0, 4.5]

car_prices_np = np.array(car_prices, dtype=np.float32).reshape(-1, 1)
car_sales_np = np.array(car_sales, dtype=np.float32).reshape(-1, 1)

car_prices_tensor = torch.from_numpy(car_prices_np)
car_sales_tensor = torch.from_numpy(car_sales_np)

plt.scatter(car_prices, car_sales)
plt.xlabel("  ($)")
plt.ylabel("  ")
plt.title("    ")
plt.grid(True)
plt.show()
```

```python
class LinearRegression(nn.Module):
    def __init__(self, input_size, output_size):
        super(LinearRegression, self).__init__()
        self.linear = nn.Linear(input_size, output_size)

    def forward(self, x):
        return self.linear(x)


input_dim = 1
output_dim =1

model = LinearRegression(input_dim, output_dim)
criterion = nn.MSELoss()

learning_rate = 0.02

loss_list = []
num_epochs = 1001

optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)
```

```python
for epoch in range(num_epochs):
    optimizer.zero_grad()
    predictions = model(car_price_tensor)
    loss = criterion(predictions, number_of_car_sell_tensor)
    loss.backward()
    optimizer.step()
    loss_list.append(loss.item())
    if epoch % 50 == 0:
        print(f"epoch {epoch}, loss = {loss.item():.4f}")
```

```python
plt.plot(range(num_epochs), loss_list)
plt.xlabel(" ")
plt.ylabel("Loss")
plt.title("   ")
plt.grid(True)
plt.show()
```
